```python
#Replacing the other null records with NA
dataset.fillna('0', inplace=True)
```

```python
dataset.dtypes
```

```python
# Checking for text values in the column before converting it to numeric datatype
dataset['Dining Rating Count'].value_counts()
```

```python
# Replacing the text values with '0'
dataset['Dining Rating Count'].replace(to_replace=['Does not offer Dining','Not enough Dining Reviews'], value='0', inplace=True)
```

```python
dataset['Delivery Rating Count'].value_counts()
```

```python
dataset['Delivery Rating Count'].replace(to_replace=['Not enough Delivery Reviews','Does not offer Delivery'], value='0', inplace=True)
```

```python
# Changing Data Type of the numerical columns
dataset['Dining Rating'] = dataset['Dining Rating'].astype('float64')
dataset['Dining Rating Count'] = dataset['Dining Rating Count'].astype('int64')
dataset['Delivery Rating'] = dataset['Delivery Rating'].astype('float64')
dataset['Delivery Rating Count'] = dataset['Delivery Rating Count'].astype('int64')
```

##Q9) Find the list of Affordable Restaurants?


```python
max_price = zomato_chndataset['Price for 2'].max()
one_fourth_price = max_price/4
one_fourth_price
```

```python
# Finding list of restaurants that have price less than and equal to 1/4th of the max price i.e Finding Cheap Restauran
aff_rest_dataset = zomato_chndataset[['Name of Restaurant', 'Price for 2', 'Top Dishes', 'Location', 'Cuisine']]
aff_rest_dataset = aff_rest_dataset[aff_rest_dataset['Price for 2'] <= 1250]
aff_rest_dataset.sort_values(by='Price for 2', inplace=True)
aff_rest_dataset
```

```python
# Finding the highest rated list of restaurants
highrate_rest_dataset = zomato_chndataset[['Name of Restaurant', 'Price for 2', 'Top Dishes', 'Location', 'Cuisine','Delivery Rating']]
highrate_rest_dataset = highrate_rest_dataset[highrate_rest_dataset['Delivery Rating'] >= 4.1]
highrate_rest_dataset.sort_values(by='Price for 2', inplace=True)
highrate_rest_dataset
```

##Q10)Find the list of most Reliable Restaurants?

```python
mean_votes = zomato_chndataset['Delivery Rating Count'].mean()
mean_votes
```

```python
# Finding list of restaurants that have Votes greater than and equal to Mean of Vote
mean_rest_dataset = zomato_chndataset[['Name of Restaurant', 'Price for 2', 'Top Dishes', 'Location', 'Cuisine', 'Delivery Rating Count']]
mean_rest_dataset = mean_rest_dataset[mean_rest_dataset['Delivery Rating Count'] >713]
mean_rest_dataset.sort_values(by='Delivery Rating Count', inplace=True)
mean_rest_dataset
```

```python
reliable_rest_dataset = pd.merge(mean_rest_dataset, highrate_rest_dataset, how='inner', on=['Name of Restaurant', 'Location'])
reliable_rest_dataset = reliable_rest_dataset[['Name of Restaurant', 'Price for 2_x', 'Top Dishes_x', 'Location', 'Cuisine_x']]
reliable_rest_dataset.rename(columns={'NAME':'Name of Restaurant', 'PRICE_x':'Price for 2', 'Top Dishes_x':'Top Dishes',
 'Location':'Location', 'Cuisine_x':'Cuisine'},inplace=True)
```
